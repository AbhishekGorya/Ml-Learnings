```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
from sklearn.datasets import make_blobs #for dataset with 5 features which will be converted into two features
```

```python
x , y = make_blobs(n_samples=500 , n_features=5 , centers=3 , cluster_std=1.5 , random_state=42)
```

```python
x
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(x_scaled)
```

```python
x_pca #with original weightage of previous features
```

```python
df_pca = pd.DataFrame(x_pca , columns=['PCA1' , 'PCA2'])
```

```python
df_pca['Labels'] = y
```

```python
sns.scatterplot(x=df_pca['PCA1'] , y=df_pca['PCA2'] , hue=df_pca['Labels'])
```

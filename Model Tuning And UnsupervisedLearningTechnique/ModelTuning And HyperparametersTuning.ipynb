```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
#Preprocessed data
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('species' , axis=1)
y = df['species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
modelKNN = KNeighborsClassifier(n_neighbors=4) #check for different values of neighbours
```

```python
modelKNN.fit(x_train , y_train)
```

```python
modelKNN.score(x_test , y_test) #if the score is 100% that means your data is overfitting for given data
```

```python
#now with SVM
```

```python
from sklearn.svm import SVC
```

```python
modelSVM = SVC(C= 30 , kernel='linear' ,gamma='auto')
modelSVM.fit(x_train , y_train)
```

```python
modelSVM.score(x_test , y_test) #again overfitted for given hyperparameters but we don't need to worry about that we are just focusing on implementation
```

```python
#so now we have seen how we change the parameters to obtain best score but you can obtain by changing parameters manually or we now we will be using GridSearchCV for finding the best Hyperparameters Set
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#perform hyperparameter tuning
```

```python
classifier = GridSearchCV((modelSVM) ,{'C':[10,20,30] , 'kernel':['rbf' , 'linear'] } , cv=5  , return_train_score=False)
```

```python
classifier.fit(x , y)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C' , 'param_kernel' ,'mean_test_score']]
```

```python
#now for KNN
```
